```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.special
import math as mt
from scipy.interpolate import interp1d
```

# Letra A

```python
# Distribuição Uniforme
N = [10, 100, 1000, 10000]

for i in range (0, np.size(N)):
    f = np.random.uniform(0,1,N[i])
    uniform, line = plt.subplots(1,2, figsize=(5,2))                              
    uniform.suptitle(f'N= {N[i]}')                                                 
    line[0].scatter([i for i in range(N[i])],f, marker='.', norm = 1.0) 
    line[1].hist(f)                                                  
    plt.show()
    
```

```python
# Distribuição Gaussiana
N = [10, 100, 1000, 10000]

for i in range (0, np.size(N)):
    f = np.random.normal(0,1,N[i])
    uniform, line = plt.subplots(1,2, figsize=(5,2))                              
    uniform.suptitle(f'N= {N[i]}')                                                 
    line[0].scatter([i for i in range(N[i])],f, marker='.', norm = 1.0) 
    line[1].hist(f)                                                  
    plt.show()
```
