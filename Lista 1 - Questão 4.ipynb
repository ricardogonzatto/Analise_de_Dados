```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.special
import math as mt
import random as rd
```

```python
def media (pontos):
    return np.sum(pontos)/(np.size(pontos))

#*variancia
def sigma (pontos, media):
    N = np.size(pontos)
    sigma_soma = [(i - media)**2 / N for i in pontos]
    return np.sum(sigma_soma)

def func_covarianc (x, y, xmedia, ymedia):
    soma = []
    for i in range(0,np.size(x)):
        soma.append(((x[i] - xmedia) * (y[i] - ymedia))  / (np.size(x) - 1))
        
    return np.sum(soma)

def func_correlat (cov, sigmax, sigmay): 
    return cov / (sigmax * sigmay)
```

```python

N = [10, 100, 1000, 10000]
soma_g = []
soma_u = []
for i in range(0, np.size(N)):
    x = np.random.uniform (0, 1, N[i])
    y = np.random.normal (0, 1, N[i])
    cov = func_covarianc(x,y, media(x), media(y))
    corr = func_correlat(cov, sigma(x,media(x)), sigma(y,media(y)))
    print(f'Covariância para N = {N[i]} : {cov}      Correlação para N = {N[i]} : {corr} ')

plt.scatter(x,y)
```
