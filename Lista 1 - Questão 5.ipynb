```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.special
import math as mt
```

```python
N =[10, 100, 1000, 10000]
#NA = np.asarray(N)
poisson_points = []
binomial_points = []
lamb = 4

for i in range(np.size(N)):
    pp = np.random.poisson(lamb, N[i])
    poisson_points.append(pp)
    p = lamb/N[i]
    b = np.random.binomial(N[i],p,N[i])
    binomial_points.append(b)
    
    plt.hist(poisson_points[i], label = f'Poisson para {N[i]} pontos')
    plt.hist(binomial_points[i], label = f'Binomial para {N[i]} pontos')
    plt.legend()
    plt.show()
```
