```python
# Distribui√ß√£o Uniforme 
N = [2,8,16,64]

for k in range (0, np.size(N)):
    pontos1=[]
    pontos2=[]
    for i in range (0,N[k]):
        pontos1.append(np.random.uniform(0,1,100))

    for j in range (0,N[k]-1):
        if j == 0:
            pontos2 = pontos1[j]+ pontos1[j+1]
        else:
            pontos2 += pontos1[j+1]
        
    
    pontos2 = [f/ np.size(N[k]) for f in pontos2]  
    g = np.size(pontos2)
    media = np.sum(pontos2)/g
    sigma_soma = [(i - media)**2 / g for i in pontos2]
    sigma = np.sqrt(np.sum(sigma_soma))
    print(sigma)
    x = np.linspace(media - 3 * sigma, media + 3 * sigma, 100)
    plt.plot(x, gauss(media, sigma, x))
    plt.hist(pontos2)
    plt.show()

```
