```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.special
import math as mt
```

```python
n = 1000
def gauss (ùúá, ùúé):
    interv_menor = ùúá - (3 * ùúé)
    interv_maior = ùúá + (3 * ùúé)
    x = np.linspace(interv_menor, interv_maior, num=n)
    y = ((1/(ùúé*np.sqrt(2*np.pi)))* np.exp(-(((x-ùúá)**2)/(2*ùúé**2))))
    return (x,y)


gaussian_points = gauss(0, 1)
plt.plot(gaussian_points[0], gaussian_points[1])
```

```python
lamb = [1,5,20,100,1000]


for i in range(np.size(lamb)):
    poisson_points = np.random.poisson(lamb[i],n)
    sigma = mt.sqrt(lamb[i])
    gaussian_points = gauss(lamb[i], sigma)
    
    plt.hist(poisson_points, bins = 100, density = True)
    plt.plot(gaussian_points[0], gaussian_points[1])
    
    plt.title(f'$\lambda$ = {lamb[i]}')
    plt.show()
```
