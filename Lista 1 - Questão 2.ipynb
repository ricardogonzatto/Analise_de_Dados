```python
n = 100
t = 1000
z = np.linspace(0,t-1, t)
var = []
media = []
xmedio = []
vmedio = []
somamedia = 0
somavar = 0
for j in range (t):
    m = np.random.normal(0,1, n)
    soma = 0
    dif = 0

    for i in range (n):
        soma += m[i]
    mediaa = soma/n
    for i in range (n):
        dif += (m[i] - mediaa)**2
    varr = dif/n
    
    somamedia += mediaa
    xmedioo = somamedia / j
    
    somavar += varr
    vmedioo = somavar / j
    
    media.append(xmedioo)
    var.append(vmedioo)
    
#print(np.size(z))
fig, ax=plt.subplots(1,2,figsize=(10,4))
ax[0].plot(z,media)
ax[1].plot(z,var)
```
